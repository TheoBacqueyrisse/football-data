```python
import matplotlib.pyplot as plt
# Create a plot
plt.figure(figsize=(10, 6))

# Plot overall_improv_list against overall_fitness_list
plt.scatter(overall_basexg, overall_fitness_list, marker='o', linestyle='-', color='b')
plt.plot(overall_basexg, overall_basexg, color="black")
# Add labels and title
plt.xlabel('Base xG')
plt.ylabel('Improved xG')
plt.title('Base xG VS Improvement')

# Optionally, add gridlines
plt.grid(True)

# Show the plot
plt.show()
```
